```python
import random
import os
import matplotlib.pyplot as plt
from PIL import Image
from typing import List
```

```python
class Config: 
    
    def __init__(self):
        self.data_dir = '../data'
        self.class_names = ['battery', 'biological', 'brown-glass', 'cardboard', 'clothes',
               'green-glass', 'metal', 'paper', 'plastic', 'shoes', 'trash', 'white-glass']
        self.file_extension = '.jpg'
```

```python
class DataVisualization:
    
    def __init__(self, data_dir: str, class_names: List[str], file_extension: str):
        self.data_dir = data_dir
        self.class_names = class_names
        self.file_extension = file_extension

    def img_for_each_class(self):
        num_classes = len(self.class_names)
        num_images_per_class = 2
        
        total_images = num_classes * num_images_per_class
        cols = 6
        rows = (total_images + cols - 1) // cols
        plt.figure(figsize=(15, 15))

        image_index = 1
        for class_name in self.class_names:
            img_dir = os.path.join(self.data_dir, class_name)
            for _ in range(num_images_per_class):
                img_name = class_name + str(random.randint(1, 50)) + self.file_extension
                img_path = os.path.join(img_dir, img_name)
                
                img = Image.open(img_path)
                ax = plt.subplot(rows, cols, image_index)
                plt.imshow(img)
                plt.title(class_name)
                plt.axis('off')
                image_index += 1

        plt.tight_layout()
        plt.show()
    
    def num_of_samples_for_each_class(self):
        dic = {}
        for class_name in self.class_names:
            img_dir = os.path.join(self.data_dir, class_name)
            num_of_samples = (len(os.listdir(img_dir)))
            dic[class_name] = num_of_samples

        names = list(dic.keys())
        values = list(dic.values())

        plt.figure(figsize=(15,10))

        plt.bar(names, values, color='maroon', width=0.6)
        plt.ylabel("Number of samples")
        plt.show()
```

```python
config = Config()
data_viz = DataVisualization(data_dir=config.data_dir, class_names=config.class_names, file_extension=config.file_extension)
```

```python
data_viz.num_of_samples_for_each_class()
```

```python
data_viz.img_for_each_class()
```
