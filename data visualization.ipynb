```python
import random
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
data_dir = './data'
class_names = ['battery', 'biological', 'brown-glass', 'cardboard', 'clothes',
               'green-glass', 'metal', 'paper', 'plastic', 'shoes', 'trash', 'white-glass']
file_extension = '.jpg'
```

```python
for class_name in class_names:
    plt.figure(figsize=(15, 15))
    img_dir = os.path.join(data_dir, class_name)
    for i in range(1, 3):
        img_name = class_name + str(random.randint(1, 50)) + file_extension
        img = Image.open(os.path.join(img_dir, img_name))
        ax = plt.subplot(4, 6, i+1)
        plt.imshow(img)
        plt.title(class_name)
        plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
dic = {}
for class_name in class_names:
    img_dir = os.path.join(data_dir, class_name)
    num_of_samples = (len(os.listdir(img_dir)))
    dic[class_name] = num_of_samples

names = list(dic.keys())
values = list(dic.values())

plt.figure(figsize=(15,10))

plt.bar(names, values, color='maroon', width=0.6)
plt.ylabel("Number of samples")
plt.show()
```
